```python
import pandas as pd
import os
import datetime as dt
import numpy as np
import json

def mark_to_numeric(mark):
    if pd.isna(mark):
        return np.nan

    mark = str(mark).strip()

    if ":" in mark:
        try:
            parts = [float(part) for part in mark.split(":")]
        except ValueError:
            return np.nan

        if len(parts) == 2:
            minutes, seconds = parts
            return minutes * 60 + seconds

        if len(parts) == 3:
            hours, minutes, seconds = parts
            return hours * 3600 + minutes * 60 + seconds

        return np.nan

    try:
        return float(mark)
    except ValueError:
        return np.nan
```

```python
try:
    df["Date"] = df.apply(lambda row: dt.datetime.strptime(str(row["Date"]), "%Y-%m-%d"), axis=1)
except ValueError:
    print(f"Error: The date is not in the correct format.")
    
```

```python
try:
    df["DOB"] = df.apply(lambda row: dt.datetime.strptime(str(row["DOB"]), "%Y-%m-%d") if pd.notna(row["DOB"]) else pd.NaT, axis=1)
except ValueError as e:
    print(f"Error: The date is not in the correct format.")
    print(e)
```

```python
df["Age"] = df.apply(lambda row: (row["Date"] - row["DOB"]).days // 365 if pd.notna(row["DOB"]) else pd.NaT, axis=1)
```

```python
import json
import pandas as pd
import pycountry

def get_country_name(country_code):
    if pd.isna(country_code):
        return None

    code = str(country_code).strip().upper()

    exceptions = {
        "URS": "RUS", "GDR": "DEU", "GER": "DEU", "ROC": "RUS",
        "NED": "NLD", "SUI": "CHE", "POR": "PRT", "MON": "MCO",
        "CRO": "HRV", "RSA": "ZAF", "NGR": "NGA",
    }

    if code in exceptions:
        return exceptions[code]

    country = pycountry.countries.get(alpha_3=code)
    if country:
        return country.alpha_3

    country = pycountry.countries.get(alpha_2=code)
    if country:
        return country.alpha_3

    return code


df_wr["Date"] = pd.to_datetime(df_wr["Date"], errors="coerce")
df_wr = df_wr.dropna(subset=["Date"]).copy()
df_wr["Year"] = df_wr["Date"].dt.year.astype(int)

df_wr["Sex"] = df_wr["Sex"].astype(str).str.strip().str.lower()
df_wr["Sex"] = df_wr["Sex"].map({"male": "M", "female": "F"})

df_wr["Country_of_Origin"] = df_wr["Nat"].apply(get_country_name)

df_cleaned = df_wr.dropna(subset=["Year", "Sex", "Country_of_Origin"]).copy()

print("Valeurs Sex uniques :", df_cleaned["Sex"].unique())
print("Nombre de lignes :", len(df_cleaned))
print("Années :", sorted(df_cleaned["Year"].unique())[:10], "...")

years = sorted(df_cleaned["Year"].unique())
sex = ["M", "F"]

final_json_structure = {}

for year in years:
    year_str = str(year)
    final_json_structure[year_str] = {"M": {}, "F": {}}

    for gender in sex:
        mask = (df_cleaned["Year"] <= year) & (df_cleaned["Sex"] == gender)
        data_until_now = df_cleaned[mask]

        if not data_until_now.empty:
            year_records_by_country = {}
            for country, group in data_until_now.groupby("Country_of_Origin"):
                year_records_by_country[country] = group[["Competitor", "Event"]].to_dict(orient="records")
            
            final_json_structure[year_str][gender] = year_records_by_country

print("Nb années dans JSON :", len(final_json_structure))
print("Première année :", next(iter(final_json_structure), None))

with open("visualizations_data/map_records/data_cumm.json", "w") as f:
    json.dump(final_json_structure, f, indent=2)
```

## Overall world records championship

```python
df_wr = df_worldrecords.copy(deep=True)
```

```python
df_wr = df_wr.dropna(subset=["Sex", "Competitor"]).copy()
df_wr["Sex"] = df_wr["Sex"].astype(str).str.strip().str.lower().map({"male": "M", "female": "F"})

df_wr["Date"] = pd.to_datetime(df_wr["Date"], errors="coerce")
df_wr["Year"] = df_wr["Date"].dt.year.astype(int)

top_athletes_data = {"M": [], "F": []}

for gender in ["M", "F"]:
    gender_data = df_wr[df_wr["Sex"] == gender]
    
    top_names = gender_data["Competitor"].value_counts().head(3).index.tolist()
    
    for name in top_names:
        athlete_records = gender_data[gender_data["Competitor"] == name]
        details = athlete_records[["Year", "Event"]].sort_values("Year").to_dict(orient="records")
        
        top_athletes_data[gender].append({
            "Competitor": name,
            "Count": len(details),
            "Details": details
        })

output_path = f"visualizations_data/top_wr_holders/data.json"

with open(output_path, "w") as f:
    json.dump(top_athletes_data, f, indent=2)
```
